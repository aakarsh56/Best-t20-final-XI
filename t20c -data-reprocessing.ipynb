```python
with open(r"C:\Users\Aakarsh Shukla\Downloads\data-analytics-project-for-beginners\t20_json_files\t20_json_files\t20_wc_batting_summary.json") as f:
    data= json.load(f)
    all_records =[]
    for rec in data:
        all_records.extend(rec['battingSummary'])
        
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
    
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: " out"if len(x) > 0 else "not_out")
df_batting.columns = df_batting.columns.str.strip()
df_batting.head(10)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
```
